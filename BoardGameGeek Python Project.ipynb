# BoardGameGeek Python Project

This uses data created by [BGG1Tool](https://boardgamegeek.com/guild/3369).
    
Data created by BGG1Tool has columns for board games such as 'Name', 'Max Players'. In addition to a 'Max Players' column which has value of the maximum number of players, there are also columns '1Player', '2Player', '3Player', etc., with possible values like 'N' (No), 'P' (Possible), or 'B' (Best).

This script creates several tables for a range of player configurations (by default 4-12). Each created table will only include the games that are possible to be played with that number of players. It then uses xlsxwriter to write each table as a sheet in a single Excel file.

For the same project done in SQL, view [this project](https://github.com/rksolana/PortfolioProjects/blob/main/BoardGameGeek%20SQL%20Project.sql).

```python
import pandas as pd
import numpy as np
# pip install xlxwriter
import xlsxwriter
# pip install inflect
import inflect
```

```python
df = pd.read_excel(r"C:\Users\Razia\Documents\result.xlsx")
pd.set_option('display.max_columns', 500)
df
```

# Add Column

Below adds a column that combines 'minplaytime' and 'maxplaytime' into a single 'playingtime' column that shows the possible range of playing time.

```python
df.insert(8, 'playingtimerange','')
for x in df.index:
    if df.loc[x, 'minplaytime'] == df.loc[x, 'maxplaytime']:
        df.loc[x, 'playingtimerange'] = str(df.loc[x, 'playingtime'])
    else:
        df.loc[x, 'playingtimerange'] = str(df.loc[x, 'minplaytime']) + '-' + str(df.loc[x, 'maxplaytime'])
df
```

# Combine Data

Below combines data from 'category' column and certain data from 'mechanic' column into a new joined column.

```python
df.insert(9, 'Categories',df['category'])
for x in df.index:
    if 'Team-Based Game' in df.loc[x, 'mechanic']:
        df.loc[x, 'Categories'] = df.loc[x, 'Categories'] + ', Team-Based Game'
    if 'Cooperative Game' in df.loc[x, 'mechanic']:
        df.loc[x, 'Categories'] = df.loc[x, 'Categories'] + ', Cooperative Game'
df
```

# Write to Excel

Below creates dataframes for a range of player counts (default 4-12), and selects only observations that fit that player count, while also renaming and selecting only certain columns.

```python
path = r"C:\Users\Razia\Documents\Board Games.xlsx"
pd.options.mode.chained_assignment = None  # default='warn'
p = inflect.engine()
writer = pd.ExcelWriter(path, engine = 'xlsxwriter')
for i in range(4,13):
    df0 = df.iloc[0:0]
    for x in df.index:
        if df.loc[x,str(i)+'player'] != 'N':
            df0.loc[len(df0.index)] = df.loc[x]
    df0_subset = df0[['name', 'maxplayers', 'playingtimerange', 'averageweight', 'Categories', 'comments_GL']]
    df0_subset.rename(columns = {'name': 'Game', 'maxplayers': 'Max Players', 'playingtimerange': 'Playing Time (minutes)', 'averageweight': 'Complexity (1-5)', 'comments_GL': 'House'}, inplace = True)
    df0_subset.to_excel(writer, sheet_name = p.number_to_words(i).capitalize(), index = False)
writer.close()
```
